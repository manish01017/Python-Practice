## Importing Modules in Python : Modules and Packages

In Python, modules and packages help organize and resuse code. Here's a comprehensive guide on how to import them.

```python
import math

math.sqrt(2)
```

```python
from math import sqrt,pi # can use  from math import * to access all functions
print(sqrt(16))
print(sqrt(8))
print(pi)
```

```python
#importing addition function from Modules.package
#restart Kernel if doesn't work

from Modules.package.math import *

print(addition(2,3))
print(substraction(2,3))
```

```python
#importing sub_packages

from Modules.package.subpackages.mult import *

print(multiplication(2,3))
```

```python
import array

arr=array.array('i',[1,2,3,4])

print(arr)
```

```python
## Data Serialization

import json
data={'name': 'Krish', 'age' :25} # dictionary
      
json_str=json.dumps(data) # changing dictionary to string
print(json_str)
print(type(json_str))


parsed_data=json.loads(json_str)
print (parsed_data)
print(type(parsed_data)) # changing string  to dictionalry
```

```python
##datetime

from datetime import datetime, timedelta

now=datetime.now()

print(now)

yesterday =now - timedelta(days=1)
print(yesterday)
```

```python
##time

import time

print(time.time())
time.sleep(3)
print(time.time())
```
